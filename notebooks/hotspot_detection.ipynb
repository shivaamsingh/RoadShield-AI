```python
import pandas as pd

df = pd.read_csv("../data/indian_roads_dataset.csv")
```

```python
coords = df[["latitude", "longitude"]]
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=10,
    random_state=42
)

df["cluster"] = kmeans.fit_predict(coords)
```

```python
df["cluster"].value_counts()
```
